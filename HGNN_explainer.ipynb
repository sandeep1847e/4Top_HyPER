```python
import os
import numpy as np
import pandas as pd
import torch
import uproot
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

from torch.nn import BCELoss, Sigmoid
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau

from torch_geometric.data import Data
from torch_geometric.loader import DataLoader

from pytorch_lightning import LightningModule, Trainer
from pytorch_lightning.loggers import TensorBoardLogger
from pytorch_lightning.callbacks import ModelCheckpoint

from sklearn.metrics import (
    roc_curve, roc_auc_score,
    confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
)

from HyPER.models import MPNNs, HyperedgeModel, EdgeLoss, CombinedLoss


OUT_DIR = "4Top_results_jan102"
PLOT_DIR = f"{OUT_DIR}/plots"
LOG_DIR  = f"{OUT_DIR}/hyperedge_logs"
CKPT_DIR = f"{OUT_DIR}/checkpoints"

for d in [OUT_DIR, PLOT_DIR, LOG_DIR, CKPT_DIR]:
    os.makedirs(d, exist_ok=True)

signal_file = "/home/sandeep/Desktop/Windows files/project rootfiles/13TeV_signal_5k.root"
background_file = "/home/sandeep/Desktop/Windows files/project rootfiles/ttbarH.root"

tree_name = "Delphes"

batch_size = 128
max_epochs = 25
delta_r_threshold = 10

def delta_r(eta1, phi1, eta2, phi2):
    dphi = (phi1 - phi2 + np.pi) % (2*np.pi) - np.pi
    deta = eta1 - eta2
    return np.sqrt(deta**2 + dphi**2)



def convert_to_hyper_data(x, eta, phi, label):
    N = x.shape[0]
    if N < 2:
        return None

    edge_index = []
    edge_attr = []
    edge_target = []

    # ----- Pairwise edges (undirected, thresholded) -----
    for i in range(N):
        for j in range(i + 1, N):
            dr = delta_r(eta[i], phi[i], eta[j], phi[j])
            if dr < delta_r_threshold:
                edge_index += [[i, j], [j, i]]
                edge_attr += [[dr], [dr]]
                edge_target += [1, 1]

    if not edge_index:
        return None

    edge_index = torch.tensor(edge_index, dtype=torch.long).T
    edge_attr_s = torch.tensor(edge_attr, dtype=torch.float)
    edge_attr_t = torch.tensor(edge_target, dtype=torch.float).view(-1, 1)

    # ----- Hyperedges (deduplicated) -----
    hyperedges = []
    for j in range(N):
        members = [
            k for k in range(N)
            if delta_r(eta[j], phi[j], eta[k], phi[k]) < delta_r_threshold
        ]
        if len(members) >= 2:
            hyperedges.append(tuple(sorted(members)))

    hyperedges = list(set(hyperedges))
    if not hyperedges:
        return None

    node_ids = np.concatenate(hyperedges)
    hedge_ids = np.repeat(
        np.arange(len(hyperedges)),
        [len(h) for h in hyperedges]
    )

    edge_index_h = torch.tensor(
        [node_ids, hedge_ids], dtype=torch.long
    )

    # ----- Global features -----
    u_s = torch.tensor(
        [[x[:, 0].sum(), x[:, 0].mean()]],
        dtype=torch.float
    )

    return Data(
        x_s=torch.tensor(x, dtype=torch.float),
        edge_index=edge_index,
        edge_attr_s=edge_attr_s,
        edge_attr_t=edge_attr_t,
        edge_index_h=edge_index_h,
        u_s=u_s,
        x_t=torch.tensor([label], dtype=torch.float),
        batch=torch.zeros(N, dtype=torch.long),
        edge_attr_s_batch=torch.zeros(edge_index.shape[1], dtype=torch.long),
        edge_index_h_batch=torch.zeros(edge_index_h.shape[1], dtype=torch.long),
    )

def load_data(file_path, label, tree_name, event_offset=0):

    branches = [
        "Jet.PT", "Jet.Eta", "Jet.Phi", "Jet.Mass",
        "Electron.PT", "Electron.Eta", "Electron.Phi",
        "Muon.PT", "Muon.Eta", "Muon.Phi",
        "MissingET.MET", "MissingET.Phi"
    ]

    tree = uproot.open(file_path)[tree_name]
    events = tree.arrays(branches, library="np")

    dataset = []

    n_events = len(events["Jet.PT"])

    for evt in range(n_events):

        # --------------------
        # Jets: [pt, eta, phi, mass]
        # --------------------
        jets = np.column_stack([
            events["Jet.PT"][evt],
            events["Jet.Eta"][evt],
            events["Jet.Phi"][evt],
            events["Jet.Mass"][evt],
        ])
        jets = np.nan_to_num(jets)

        # --------------------
        # Leptons (e + mu): [pt, eta, phi, mass=0]
        # --------------------
        lep_pt  = np.concatenate([events["Electron.PT"][evt],  events["Muon.PT"][evt]])
        lep_eta = np.concatenate([events["Electron.Eta"][evt], events["Muon.Eta"][evt]])
        lep_phi = np.concatenate([events["Electron.Phi"][evt], events["Muon.Phi"][evt]])

        if len(lep_pt) > 0:
            leptons = np.column_stack([
                lep_pt,
                lep_eta,
                lep_phi,
                np.zeros(len(lep_pt))
            ])
            leptons = np.nan_to_num(leptons)
        else:
            leptons = np.empty((0, 4))

        # --------------------
        # MET as pseudo-node (optional but consistent)
        # --------------------
        met_pt  = events["MissingET.MET"][evt][0]
        met_phi = events["MissingET.Phi"][evt][0]

        met = np.array([[met_pt, 0.0, met_phi, 0.0]])
        met = np.nan_to_num(met)

        # --------------------
        # Stack all nodes
        # --------------------
        x = np.vstack([jets, leptons, met])

        eta = x[:, 1]
        phi = x[:, 2]

        data = convert_to_hyper_data(x, eta, phi, label)
        if data is None:
            continue

        # --------------------
        # Event-level metadata
        # --------------------
        jets_pt_sorted = np.sort(jets[:, 0])[::-1]

        data.event_id = event_offset + evt   # SAFE when merging datasets
        data.n_jets = jets.shape[0]
        data.lead_jet_pt = float(jets_pt_sorted[0]) if len(jets_pt_sorted) > 0 else np.nan
        data.sublead_jet_pt = float(jets_pt_sorted[1]) if len(jets_pt_sorted) > 1 else np.nan
        data.lead_lep_pt = float(np.max(lep_pt)) if len(lep_pt) > 0 else np.nan

        dataset.append(data)

    return dataset

class HyPERModel(LightningModule):

    def __init__(self):
        super().__init__()
        self.train_loss_epoch = []
        self.val_loss_epoch = []


        self.model = MPNNs(
            node_in_channels=4,
            edge_in_channels=1,
            global_in_channels=2,
            node_out_channels=128,
            edge_out_channels=1,
            global_out_channels=128,
            message_feats=128,
            dropout=0.2,
            activation=Sigmoid(),
            p_out="edge"
        )

        self.hyper = HyperedgeModel(
            node_in_channels=128,
            node_out_channels=1,
            global_in_channels=128,
            message_feats=128,
            dropout=0.2
        )

        self.edge_loss_fn = BCELoss(reduction="none")
        self.hyper_loss_fn = BCELoss()
        self.alpha = 0.4

        self.event_pred_log = []
        self.hyperedge_log = []

    def forward(self, data):
        x, edge_score, u = self.model(
            data.x_s, data.edge_index,
            data.edge_attr_s, data.u_s,
            data.batch
        )

        x_hat, batch_hyper = self.hyper(
            x, u, data.batch,
            data.edge_index_h,
            data.edge_index_h_batch,
            3
        )
        return x_hat, batch_hyper, edge_score

    def training_step(self, batch, _):
        x_hat, batch_hyper, edge_score = self(batch)

        x_evt = torch.zeros_like(batch.x_t)
        for i in range(len(batch.x_t)):
            mask = (batch_hyper == i)
            if mask.any():
                x_evt[i] = x_hat[mask].mean()
            else:
                x_evt[i] = 0.0

        loss_h = self.hyper_loss_fn(x_evt, batch.x_t)

        loss_e = EdgeLoss(
            edge_score.view(-1),
            batch.edge_attr_t.view(-1),
            batch.edge_attr_s_batch,
            self.edge_loss_fn,
            reduction="mean"
        )

        loss = CombinedLoss(loss_e, loss_h, self.alpha, "mean")
        self.log("train_loss", loss, on_epoch=True, prog_bar=True)
        self.train_loss_epoch.append(loss.detach().cpu().item())

        return loss
    def validation_step(self, batch, _):
        x_hat, batch_hyper, edge_score = self(batch)
        edge_score = edge_score.view(-1)

        # -----------------------------
        # Event-level prediction first
        # -----------------------------
        x_evt = torch.zeros_like(batch.x_t)
        for i in range(len(batch.x_t)):
            mask = (batch_hyper == i)
            x_evt[i] = x_hat[mask].mean() if mask.any() else 0.0

        val_loss = self.hyper_loss_fn(x_evt, batch.x_t)
        self.log("val_loss", val_loss, on_epoch=True, prog_bar=True)
        self.val_loss_epoch.append(val_loss.detach().cpu().item())

        # -----------------------------
        # Event-level detailed logging
        # -----------------------------
        for i in range(len(batch.x_t)):
            evt_id = int(batch.event_id[i])
            n_jets_evt = int(batch.n_jets[i])

            evt_mask = (batch.batch == i)
            evt_nodes = batch.x_s[evt_mask]

            # total nodes in this event
            n_nodes_evt = evt_nodes.shape[0]

            # Leptons = everything between jets and MET
            n_leptons = n_nodes_evt - n_jets_evt - 1

            # Jets
            jet_pts  = evt_nodes[:n_jets_evt, 0].cpu().tolist()
            jet_etas = evt_nodes[:n_jets_evt, 1].cpu().tolist()
            jet_phis = evt_nodes[:n_jets_evt, 2].cpu().tolist()

            # Lepton pT
            lepton_pts = (
                evt_nodes[n_jets_evt:-1, 0].cpu().tolist()
                if n_leptons > 0 else []
            )

            # MET (last node)
            met_pt = float(evt_nodes[-1, 0].item())

            # Hyperedge count
            n_hyperedges_evt = int((batch_hyper == i).sum())

            self.event_pred_log.append({
                "event_id": evt_id,
                "true_label": int(batch.x_t[i]),
                "pred_score": float(x_evt[i]),
                "pred_label": int(x_evt[i] > 0.5),

                "jet_pts": jet_pts,
                "jet_etas": jet_etas,
                "jet_phis": jet_phis,
                "lepton_pts": lepton_pts,
                "n_leptons": n_leptons,
                "met_pt": met_pt,
                "n_hyperedges": n_hyperedges_evt,
            })


        # Hyperedge-level logging

        for h in range(len(x_hat)):

            evt = int(batch_hyper[h])
            evt_id = int(batch.event_id[evt])
            n_jets_evt = int(batch.n_jets[evt])

            # nodes in this hyperedge (GLOBAL node ids)
            nodes = batch.edge_index_h[0][batch.edge_index_h[1] == h]
            if nodes.numel() == 0:
                continue

            # event-local node ids
            evt_node_ids = torch.where(batch.batch == evt)[0]

            # map global → local indices
            local_nodes = torch.searchsorted(evt_node_ids, nodes)

            # jets are ALWAYS first in local indexing
            jets = local_nodes[local_nodes < n_jets_evt]

            # HARD PHYSICS CUT — THIS MUST BE HERE
            if jets.numel() < 1:
                continue

            leptons = local_nodes[
                (local_nodes >= n_jets_evt) &
                (local_nodes < (evt_node_ids.numel() - 1))
            ]

            # ---- geometry ----
            etas = batch.x_s[evt_node_ids[jets], 1].cpu().numpy()
            phis = batch.x_s[evt_node_ids[jets], 2].cpu().numpy()

            drs = [
                delta_r(etas[i], phis[i], etas[j], phis[j])
                for i in range(len(jets))
                for j in range(i + 1, len(jets))
            ]

            # ---- edge-level scores ----
            edge_scores = []
            for i in nodes:
                for j in nodes:
                    if i == j:
                        continue
                    mask = (batch.edge_index[0] == i) & (batch.edge_index[1] == j)
                    if mask.any():
                        edge_scores.append(float(edge_score[mask][0]))

            self.hyperedge_log.append({
                "event_id": evt_id,
                "label": int(batch.x_t[evt]),
                "hyperedge_score": float(x_hat[h]),

                "n_jets_hyperedge": int(jets.numel()),
                "n_leptons_hyperedge": int(leptons.numel()),
                "hyperedge_size_total": int(nodes.numel()),

                "mean_dr": float(np.mean(drs)),
                "max_dr": float(np.max(drs)),
                "edge_scores": edge_scores,
            })

    def on_validation_epoch_end(self):
        if not self.event_pred_log:
            return

        df = pd.DataFrame(self.event_pred_log)
        y_true = df.true_label.values
        y_score = df.pred_score.values

        if len(np.unique(y_true)) > 1:
            auc = roc_auc_score(y_true, y_score)
            fpr, tpr, _ = roc_curve(y_true, y_score)

            plt.figure()
            plt.plot(fpr, tpr, label=f"AUC={auc:.3f}")
            plt.plot([0, 1], [0, 1], "--")
            plt.xlabel("False Positive Rate")
            plt.ylabel("True Positive Rate")
            plt.title(f"ROC Curve ")
            plt.legend()
            plt.show()
            plt.savefig(f"{PLOT_DIR}/roc_epoch_{self.current_epoch}.png")
            plt.close()



            # ---- Confusion Matrix ----
            y_pred = (y_score > 0.5).astype(int)
            cm = confusion_matrix(y_true, y_pred)
            cm_norm = cm / cm.sum(axis=1, keepdims=True)

            plt.figure(figsize=(5,4))
            sns.heatmap(
                cm_norm,
                annot=True,
                fmt=".2f",
                cmap="Blues",
                xticklabels=["Bkg", "Sig"],
                yticklabels=["Bkg", "Sig"]
            )
            plt.xlabel("Predicted")
            plt.ylabel("True")
            plt.title(f"Confusion Matrix (Epoch {max_epochs})")
            plt.show()
            plt.close()

            # ---- Metrics ----
            acc  = accuracy_score(y_true, y_pred)
            prec = precision_score(y_true, y_pred)
            rec  = recall_score(y_true, y_pred)
            f1   = f1_score(y_true, y_pred)

            
            print(f"AUC: {auc:.4f}\n"),print(f"Accuracy: {acc:.4f}\n",f"Precision : {prec:.4f}\n",f"Recall: {rec:.4f}\n",f"F1-score  : {f1:.4f}\n")

            


        joblib.dump(
            self.hyperedge_log,
            f"{LOG_DIR}/hyperedges_epoch_{self.current_epoch}.pkl"
        )
        joblib.dump(
            self.event_pred_log,
            f"{LOG_DIR}/events_epoch_{self.current_epoch}.pkl"
        )
    # ---- Loss curve ----
        plt.figure(figsize=(7,5))
        plt.plot(self.train_loss_epoch, label="Train Loss")
        plt.plot(self.val_loss_epoch, label="Val Loss")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.grid()
        plt.title("Training & Validation Loss")
        plt.show()
        plt.close()

        self.hyperedge_log.clear()
        self.event_pred_log.clear()

    def configure_optimizers(self):
        opt = Adam(self.parameters(), lr=1e-3)
        sch = ReduceLROnPlateau(opt, mode="min", patience=5)
        return {"optimizer": opt, "lr_scheduler": sch, "monitor": "val_loss"}

signal = load_data(signal_file, 1, tree_name)
background = load_data(background_file, 0, tree_name)

dataset = signal + background
np.random.shuffle(dataset)

split = int(0.6 * len(dataset))
train_loader = DataLoader(dataset[:split], batch_size=batch_size, shuffle=True)
val_loader = DataLoader(dataset[split:], batch_size=batch_size)

trainer = Trainer(
    max_epochs=max_epochs,
    accelerator="gpu",
    devices=1,
    logger=TensorBoardLogger(OUT_DIR, name="lightning"),
    callbacks=[ModelCheckpoint(
        dirpath=CKPT_DIR,
        monitor="val_loss",
        save_top_k=1
    )],
    num_sanity_val_steps=0
)

trainer.fit(HyPERModel(), train_loader, val_loader)
```
